          **Data Understanding**

```python
#importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("title.basics.csv")
df2 = pd.read_csv("bom.movie_gross.csv")
df3 = pd.read_csv("title.ratings.csv")
```

```python
df1.head()
```

```python
df2.head()
```

```python
df3.head()
```

```python
to_concat = [df1, df2, df3]
movie_df = pd.concat(to_concat)
```

```python
movie_df.isna().sum()
```

     **Data cleaning**

```python
# Fill missing values in 'runtime_minutes' with the mean
mean= movie_df['runtime_minutes'].mean()
movie_df['runtime_minutes'].fillna(mean, inplace=True)
```

```python
# Fill missing values in 'averagerating' with 0
movie_df['averagerating'].fillna(0.00000, inplace=True)
```

```python
# Fill missing values in 'genres' with  'Unknown'
movie_df['genres'].fillna('Unknown', inplace=True)
```

```python
# Fill missing values in 'numvotes' with 0
movie_df['numvotes'].fillna(0.00000, inplace=True)
```

```python
columns_to_drop=['title','studio','domestic_gross','foreign_gross','year']
movie_df=movie_df.drop(columns = columns_to_drop)
print (movie_df)
```

```python
movie_df.isna().sum()
```

```python
movie_df.describe()
```

    ** Analysis**

    Rating of the movies

  'averagerating' is a quantitative representation of the collective opinions and preferences of viewers, serving as a crucial metric for evaluating the perceived  popularity of movies

```python
#plot a bargraph to compare the averagerating
plt.figure(figsize=(10, 6))
movie_df['averagerating'].value_counts().sort_index().plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Number of Movie Releases by averagerating')
plt.xlabel('averagerating')
plt.ylabel('Number of Movies')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**GENRES ANALYSIS**

This visualization allows for a quick understanding of the distribution of movie genres , providing insights into which genres are more prevalent and potentially more popular among the movies.Documentary and Drama are the most popular.

```python
#plot a bargraph to compare genre
plt.figure(figsize=(14, 8))
movie_df['genres'].str.split(',', expand=True).stack().value_counts().plot(kind='bar', color='lightgreen', edgecolor='black')

plt.title('Distribution of Movie Genres')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Start year and Runtime minutes**
The runtime of majority of the movies is 85-90 minutes



```python
#plot a bargraph comparing start year and runtime minutes
plt.figure(figsize=(14, 8))

movie_df_grouped = movie_df.groupby('start_year')['runtime_minutes'].mean().reset_index()

plt.bar(movie_df_grouped['start_year'], movie_df_grouped['runtime_minutes'], color='orange', label='Average Runtime')
plt.xlabel('Start Year')
plt.ylabel('Average Runtime (minutes)')
plt.title('Comparison of Average Runtime by Start Year')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# comparing genres and averagerating
x = 'genres'
y = 'averagerating'

# Create a bar graph using Seaborn
plt.figure(figsize=(14, 8))
sns.barplot(x=x, y=y, data=movie_df, errorbar=None, palette='viridis')
plt.xlabel('Genres')
plt.ylabel('Average Rating')
plt.title('Bar Graph: Average Rating by Genres')
plt.xticks(rotation=100, ha='left')
plt.show()
```


**Start_Year**

 Each bar represents a range of start years, and the height of the bar indicates the frequency or count of movies that started in that particular year range.

```python
# Create a histogram for 'start_year'
plt.figure(figsize=(10, 6))
plt.hist(movie_df['start_year'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Start Year')
plt.xlabel('Start Year')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

  
  Average rating over time
  
  This line plot provides a visual representation of how the average rating of movies has changed over the years. Each point on the line represents the average rating for a specific year, and the line connects these points to show the overall trend

```python
# Line plot: Average Rating over Time
plt.figure(figsize=(14, 8))
sns.lineplot(x='start_year', y='averagerating', data=movie_df, errorbar=None, marker='o', color='blue')
plt.xlabel('Start Year')
plt.ylabel('Average Rating')
plt.title('Line Plot: Average Rating Over Time')
plt.xticks(rotation=45, ha='right')
plt.show()
```
